```python
from neo4j import GraphDatabase
import logging
from neo4j.exceptions import ServiceUnavailable
```

```python
class App:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        # Don't forget to close the driver connection when you are finished with it
        self.driver.close()

    def create_friendship(self, bugid,types,description,product, component,status,resolution,asignee, priority,severity,classification):
        with self.driver.session() as session:
            # Write transactions allow the driver to handle retries and transient errors
            result = session.write_transaction(
                self._create_and_return_friendship, bugid,types,description,product, component,status,resolution,asignee, priority,severity,classification)
            for row in result:
                print("Created Node between: {p1}, {p2}".format(p1=row['p1'], p2=row['p2']))
           
          
    

    @staticmethod
    def _create_and_return_friendship(tx,bugid,types,description,product, component,status,resolution,asignee, priority,severity,classification):
        # To learn more about the Cypher syntax, see https://neo4j.com/docs/cypher-manual/current/
        # The Reference Card is also a good resource for keywords https://neo4j.com/docs/cypher-refcard/current/
        query = (
            "MERGE (p1:bugid { name: $bugid }) "
            "MERGE (p2:types { name: $types }) "
            "MERGE (p3:description { name: $description }) "
            "MERGE (p4:product { name: $product }) "
            "MERGE (p5:component { name: $component }) "
            "MERGE (p6:status { name: $dstatus }) "
            "MERGE (p7:resolution { name: $resolution }) "
            "MERGE (p8:asignee { name: $asignee }) "
            "MERGE (p9:priority { name: $pirority }) "
            "MERGE (p10:severity{ name: $severity }) "
            "MERGE (p11:classification{ name: $classification }) "
        )

        relationQuery=(
            "CREATE (p1)-[:isA]->(p2) "
            "CREATE (p1)-[:are]->(p3) "
            "CREATE (p1)-[:product]->(p4) "
            "CREATE (p1)-[:have]->(p5) "
            "CREATE (p1)-[:status]->(p6) "
            "CREATE (p1)-[:res]->(p7) "
            "CREATE (p1)-[:solvedBy]->(p8) "
            "CREATE (p1)-[:pri]->(p9) "
            "CREATE (p1)-[:seve]->(p10) "
            "CREATE (p1)-[:classified]->(p11) "
        )

        returnQueryVar="RETURN p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11"

        query+=relationQuery
        result = tx.run(query, bugid=bugid, types=types,description=description,product=product, component=component,status=status,resolution=resolution,asignee=asignee, priority=priority,severity=severity,classification=classification)
        query+=returnQueryVar
      
        
        try:
            return [{"p1": row["p1"]["name"], "p2": row["p2"]["name"]}
                    for row in result]
        # Capture any errors along with the query and data for traceability
        except ServiceUnavailable as exception:
            logging.error("{query} raised an error: \n {exception}".format(
                query=query, exception=exception))
            raise
        
#     def find_person(self, person_name):
#         with self.driver.session() as session:
#             result = session.read_transaction(self._find_and_return_person, bugid)
#             for row in result:
#                 print("Found person: {row}".format(row=row))

    @staticmethod
#     def _find_and_return_person(tx, person_name):
#         query = (
#             "MATCH (p:Person) "
#             "WHERE p.name = $person_name "
#             "RETURN p.name AS name"
#         )
#         result = tx.run(query, person_name=person_name)
#         return [row["name"] for row in result]


if ___name__ == "__main__":
    # Aura queries use an encrypted connection using the "neo4j+s" URI scheme
    uri = "neo4j://localhost:7687"
    user = "sanjoligo6mar@gmail.com"
    password = "4GttC5HCn9!SwBvg"
    app = App(uri, user, password)
    for index, rows in bugs.iterrows():
        bugId= bugs['Bug ID']
        types= bugs['Type']
        description= bugs['Summary']
        product=bugs['Product']
        component= bugs['Component']
        status= bugs['Status']
        resolution= bugs['Resolution']
        assignee= bugs['Assignee']
        priority= bugs['Priority']
        severity= bugs['Severity']
        classification=bugs['Classification']
        app.create_friendship( bugid,types,description,product, component,status,resolution,asignee, priority,severity,classification)
    app.close()
```
